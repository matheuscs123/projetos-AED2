```python
#questão 1 - capitulo 2

import pandas as pd
#o exercicio pede para calcular a probabilidade e obter os diferentes resultados diferentes, ao johar duas moedas com porbabilodade de cara(H) oo coroa(T)

# valores p1(H) e p2(H) da tabela.
p1 = [0.5, 0.6, 0.4, 0.1, 0.3]
p2 = [0.5, 0.7, 0.8, 0.2, 0.4]

# Criando uma tabela que armazena os resultados
resultados = {
    'p1(H)': p1,
    'p2(H)': p2,
    'HH': [],
    'HT': [],
    'TH': [],
    'TT': []
}

# Calculando as probabilidades para cada combinação
#sendo primeira moeda dar cara = P1(i)
#sendo a segunda moeda dar cora = p2[i] 
#sendo a primeira moeda dar coroa = (1-p1(i))
#senod a segunda moeda dar coroa = (1-p2[i]) 
for i in range(len(p1)):
    #possibilidades
    prob_HH = p1[i] * p2[i]  #cara - cara
  
    prob_HT = p1[i] * (1 - p2[i]) #cara - coroa
    
    prob_TH = (1 - p1[i]) * p2[i] #coroa - cara
    
    prob_TT = (1 - p1[i]) * (1 - p2[i]) #coroa - coroa
    
    resultados['HH'].append(prob_HH)
    resultados['HT'].append(prob_HT)
    resultados['TH'].append(prob_TH)
    resultados['TT'].append(prob_TT)

# Convertendo para um DataFrame para exibir em formato de tabela
df_resultados = pd.DataFrame(resultados)

# mostrando os resultados possíveis 

df_resultados
#Através deste exercício, aprendemos a calcular as probabilidades de diferentes resultados ao jogar duas moedas com probabilidades distintas.
#O uso de um DataFrame do pandas facilita a visualização e análise dos dados obtidos.

```

```python
# queatão 2 - capitulo 2
# Definindo os valores iniciais
P_S = 0.60  # probabilidade de um e-mail ser spam
P_NS = 0.40  # probabilidade de um e-mail não ser spam
P_F_given_S = 0.98  # probabilidade do filtro sinalizar spam dado que é spam
P_F_given_NS = 0.02  # probabilidade do filtro sinalizar spam dado que não é spam

# Número de e-mails sinalizados como spam
n_spam_flagged = 963

#Cálculo de P(F):
#P(F) é a probabilidade de o filtro marcar um e-mail como spam. Isso pode acontecer em dois casos:
#O e-mail realmente é spam e o filtro marca como spam.
#O e-mail não é spam e o filtro marca como spam.

# Usando a fórmula da probabilidade total para calcular P(F)
#
P_F = P_F_given_S * P_S + P_F_given_NS * P_NS

# Usando o Teorema de Bayes para calcular P(NS|F)
P_NS_given_F = (P_F_given_NS * P_NS) / P_F #substituinod os valores iniciais

# Calculando o número de falsos positivos (e-mails não spam na caixa de spam)
falsos_positivos = P_NS_given_F * n_spam_flagged

# Resultados
P_NS_given_F, falsos_positivos
#O resultado correto, considerando o Teorema de Bayes e os dados fornecidos, é:
#A probabilidade de um e-mail sinalizado como spam não ser realmente spam (P(NS|F)) é aproximadamente 1,34%.
#O número esperado de falsos positivos (e-mails que não são spam, mas estão na caixa de spam) é aproximadamente 13 e-mails.
```

```python
#questão 3 - capitulo 2
# Tabela de mapeamento entre letras antigas e novas
mapeamento = {
    'A': 'T', 'B': 'U', 'C': 'V', 'D': 'W', 'E': 'X', 'F': 'Y', 'G': 'Z', 'H': 'A', 'I': 'B', 'J': 'C',
    'K': 'D', 'L': 'E', 'M': 'F', 'N': 'G', 'O': 'H', 'P': 'I', 'Q': 'J', 'R': 'K', 'S': 'L', 'T': 'M',
    'U': 'N', 'V': 'O', 'W': 'P', 'X': 'Q', 'Y': 'R', 'Z': 'S'
}

# Função para aplicar a cifra de César (deslocamento de 7) 3 vezes
def aplicar_cifra(texto, mapeamento, etapas=3):
    for _ in range(etapas):
        novo_texto = ''
        for letra in texto:
            if letra in mapeamento:  # Se a letra está no mapeamento, aplica a substituição
                novo_texto += mapeamento[letra]
            else:
                novo_texto += letra  # Mantém espaços e outros caracteres intactos
        texto = novo_texto
    return texto

# Sequência codificada
texto_codificado = "OCZ XJMMZXO VINRZM"

# Decodificação aplicando o deslocamento 7 três vezes
texto_decodificado = aplicar_cifra(texto_codificado, mapeamento)

# Exibir o texto decodificado
print("Texto decodificado:", texto_decodificado)
```

```python
#questão 4 - capitulo 2
#Teremos duas listas: uma para os resultados e outra para as probabilidades (p).
#Função de Massa de Probabilidade:
#Isso será representado como um gráfico de barras, onde cada barra representa a probabilidade de um resultado específico.
#Vamos somar as probabilidades de cada valor para formar a distribuição cumulativa.
import matplotlib.pyplot as plt
import numpy as np
# tem uma série de resultados (neste caso, de 1 a 8) e suas respectivas probabilidades associadas (valores de 𝑝)
#Cada probabilidade representa a chance de um resultado específico ocorrer em um experimento.

# Dados
resultados = [1, 2, 3, 4, 5, 6, 7, 8]  #A lista resultados contém os valores de 1 a 8.
probabilidades = [0.1, 0.15, 0.2, 0.21, 0.17, 0.09, 0.06, 0.02]  #A lista probabilidades contém as probabilidades associadas a cada resultado.

# Função de Massa de Probabilidade (PMF) #A PMF mostra a probabilidade de cada resultado individual. Por exemplo:
#O resultado 1 tem uma probabilidade de 0,1, o que significa que, em 100 experimentos,
#espera-se que o resultado 1 ocorra aproximadamente 10 vezes.
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.bar(resultados, probabilidades, color='blue', alpha=0.7)
plt.title('Função de Massa de Probabilidade')
plt.xlabel('Resultado')
plt.ylabel('Probabilidade')

# Distribuição Cumulativa
probabilidades_cumulativas = np.cumsum(probabilidades)

plt.subplot(1, 2, 2)
plt.step(resultados, probabilidades_cumulativas, where='mid', color='green', alpha=0.7)
plt.title('Distribuição Cumulativa')
plt.xlabel('Resultado')
plt.ylabel('Probabilidade Cumulativa')

plt.tight_layout()
plt.show()
```

```python
#questão 5 - capitulo 2
#Para determinar quantos bits precisamos para codificar independemente 𝑣1 e 𝑣2, e quanto economizaríamos na codificação de 𝑣1, 
#se soubéssemos 𝑣2, utiliza-se o conceito de entropia.
#bits pra codificar 𝑣1 e 𝑣2 independente:
import numpy as np

# Distribuições de probabilidade de exemplo
p_v1 = np.array([0.4, 0.6])  # Probabilidades de v1
p_v2 = np.array([0.5, 0.5])  # Probabilidades de v2

# Função para calcular a entropia
def entropy(p):
    return -np.sum(p * np.log2(p))

# Entropias individuais
H_v1 = entropy(p_v1)
H_v2 = entropy(p_v2)

# Entropia conjunta (assumindo independência)
H_v1_v2 = H_v1 + H_v2

# Entropia condicional H(v1 | v2)
# Para simplificação, assumimos que a distribuição condicional é igual à distribuição marginal de v1
H_v1_given_v2 = H_v1

# Economia na codificação de v1 sabendo v2
economia = H_v1 - H_v1_given_v2

print(f"Entropia de v1: {H_v1:.2f} bits")
print(f"Entropia de v2: {H_v2:.2f} bits")
print(f"Entropia conjunta H(v1, v2): {H_v1_v2:.2f} bits")
print(f"Entropia condicional H(v1 | v2): {H_v1_given_v2:.2f} bits")
print(f"Economia na codificação de v1 sabendo v2: {economia:.2f} bits")




```

```python
#questão 1 - capitulo 3
# Definindo o gráfico como uma lista de adjacência
#O gráfico é representado como um dicionário onde as chaves são os vértices e os valores são listas de vértices adjacentes.
#O número de vértices é simplesmente o número de chaves no dicionário.
#O número de arestas é calculado somando o comprimento de todas as listas de adjacência e dividindo por 2 (para evitar contar cada aresta duas vezes).
import networkx as nx

# Criando um grafo vazio
G = nx.Graph()

# Adicionando os nós (vértices)
# Como não temos informações específicas sobre os nós, podemos usar números como identificadores
nodes = [1, 2, 3, 4, 5]
G.add_nodes_from(nodes)

# Adicionando as arestas (conexões)
# Observando a imagem, podemos adicionar as arestas da seguinte forma:
edges = [(1, 2), (1, 3), (2, 3), (3, 4), (4, 5)]
G.add_edges_from(edges)

# Calculando |V| e |E|
num_vertices = len(G.nodes)
num_arestas = len(G.edges)

print("Número de vértices (|V|):", num_vertices)
print("Número de arestas (|E|):", num_arestas)

# Visualizando o grafo (opcional)
import matplotlib.pyplot as plt
nx.draw(G, with_labels=True)
plt.show()
```

```python
#questao 2 - capitulo 4
import networkx as nx
import matplotlib.pyplot as plt

# Criar um grafo direcionado
G = nx.DiGraph()

# Adicionando arestas de acoplamento entre camadas ("C")
edges_between_layers = [
    (1, 2, 'C'),
    (1, 3, 'C'),
    (1, 4, 'C'),
    (5, 6, 'C'),
    (5, 7, 'C'),
    (5, 8, 'C'),
    (9, 10, 'C'),
    (9, 11, 'C'),
    (9, 12, 'C'),
    (13, 14, 'C'),
    (13, 15, 'C'),
    (13, 16, 'C'),
]

# Adicionando arestas dentro das camadas
edges_within_layers = [
    (1, 5, 1),
    (1, 9, 1),
    (1, 13, 1),
    (5, 9, 1),
    (5, 13, 1),
    (9, 13, 1),
    (2, 6, 2),
    (6, 10, 2),
    (10, 14, 2),
    (14, 2, 2),
    (3, 7, 3),
    (7, 11, 3),
    (11, 15, 3),
    (4, 8, 4),
    (12, 16, 4),
]

# Adicionando arestas ao grafo
for u, v, layer in edges_between_layers:
    G.add_edge(u, v, type=layer)

for u, v, layer in edges_within_layers:
    G.add_edge(u, v, type=layer)

# Função para verificar se existe um acoplamento estrela
def has_star_coupling(graph):
    for node in graph.nodes():
        neighbors = list(graph.successors(node))
        if len(neighbors) > 1 and all(graph.has_edge(node, n) for n in neighbors):
            return True
    return False

# Função para verificar se existe um clique
def has_clique(graph):
    for node in graph.nodes():
        neighbors = list(graph.successors(node))
        if len(neighbors) > 1:
            # Verifica se todos os vizinhos estão conectados entre si
            if all(graph.has_edge(n1, n2) for i, n1 in enumerate(neighbors) for n2 in neighbors[i+1:]):
                return True
    return False

# Função para verificar se existe uma cadeia
def has_chain(graph):
    for node in graph.nodes():
        if len(list(graph.successors(node))) == 2:  # Cada nó deve conectar a dois outros nós
            return True
    return False

# Verificação das propriedades da rede
star_coupling = has_star_coupling(G)
clique = has_clique(G)
chain = has_chain(G)

# Exibindo os resultados
print("A rede possui um acoplamento estrela?", star_coupling)
print("A rede possui um clique?", clique)
print("A rede possui uma cadeia?", chain)

# Visualizando a rede
plt.figure(figsize=(12, 8))
pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True, arrows=True)
edge_labels = nx.get_edge_attributes(G, 'type')
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels)
plt.title("Rede Multicamada")
plt.show()
```

```python
#questão 3 - capitulo 4
import networkx as nx
import matplotlib.pyplot as plt

# Definindo a rede hipergrafo com as hiperarestas
hyperedges = [
    [1, 2, 3, 4, 5, 6],
    [2, 4, 7, 8],
    [4, 5, 6, 7, 10],
    [6, 9, 10, 11, 12],
    [11, 13, 14, 15],
    [11, 12, 13],
    [14, 15, 16, 17, 18],
    [15, 17, 19, 20],
    [21, 22, 23],
    [23, 24],
    [24, 25, 26],
    [25, 27, 28, 29]
]

# Criando a rede unipartida
G_unipartite = nx.Graph()

# Adicionando arestas entre todos os nós de cada hiperaresta
for hyperedge in hyperedges:
    for i in range(len(hyperedge)):
        for j in range(i + 1, len(hyperedge)):
            G_unipartite.add_edge(hyperedge[i], hyperedge[j])

# Visualizando a rede unipartida
plt.figure(figsize=(10, 6))
plt.title("Rede Unipartida")
nx.draw(G_unipartite, with_labels=True, node_color='skyblue', font_size=10, font_weight='bold')
plt.show()

# Criando a rede bipartida
G_bipartite = nx.Graph()

# Adicionando nós para as hiperarestas
for index, hyperedge in enumerate(hyperedges):
    hyperedge_node = f'H{index + 1}'  # Nome do nó da hiperaresta
    G_bipartite.add_node(hyperedge_node)  # Adicionando o nó da hiperaresta
    for node in hyperedge:
        G_bipartite.add_edge(hyperedge_node, node)  # Conectando a hiperaresta aos seus nós

# Visualizando a rede bipartida
plt.figure(figsize=(10, 6))
plt.title("Rede Bipartida")
nx.draw(G_bipartite, with_labels=True, node_color='salmon', font_size=10, font_weight='bold')
plt.show()
```

```python
#questão 4 - capitulo 4
import networkx as nx
import matplotlib.pyplot as plt

# Definindo as arestas da rede dinâmica com os períodos de tempo
edges_data = [
    (1, 2, 1, 4),
    (1, 2, 7, 9),
    (1, 3, 3, 7),
    (1, 4, 2, 4),
    (1, 5, 1, 9),
    (2, 3, 1, 2),
    (2, 4, 1, 3),
    (2, 4, 6, 7),
    (2, 4, 9, 9),
    (3, 5, 5, 7),
    (3, 6, 1, 3),
    (3, 6, 8, 9),
    (4, 7, 1, 9),
    (5, 8, 1, 2),
    (5, 8, 4, 7),
    (5, 8, 9, 9),
    (5, 9, 1, 5),
    (5, 10, 8, 9),
    (5, 11, 7, 8),
    (6, 12, 4, 7),
    (6, 13, 2, 3),
    (7, 12, 2, 6)
]

# Criar um grafo dinâmico
dynamic_graph = nx.Graph()

# Definindo uma função para agregar as arestas usando uma janela disjunta de tamanho 3
def add_edges_with_window(edges, window_size):
    # Criar um conjunto de períodos de tempo
    time_slots = set()
    for _, _, start, end in edges:
        time_slots.update(range(start, end + 1))

    # Ordenar os períodos de tempo
    sorted_time_slots = sorted(time_slots)

    # Iterar sobre os períodos de tempo usando uma janela disjunta
    for start_time in range(min(sorted_time_slots), max(sorted_time_slots) + 1, window_size):
        end_time = start_time + window_size - 1
        present_edges = set()
        
        # Verificar quais arestas estão presentes nesse intervalo
        for u, v, start, end in edges:
            if start <= end_time and end >= start_time:
                present_edges.add((u, v))
        
        # Adicionar as arestas presentes ao grafo dinâmico
        dynamic_graph.add_edges_from(present_edges)

# Chamar a função para agregar as arestas
add_edges_with_window(edges_data, window_size=3)

# Visualizando o grafo dinâmico agregado
plt.figure(figsize=(10, 6))
plt.title("Grafo Dinâmico Agregado (Janela de Tamanho 3)")
nx.draw(dynamic_graph, with_labels=True, node_color='lightblue', font_size=10, font_weight='bold')
plt.show()
```
